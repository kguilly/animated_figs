
# US Counties

Demonstrate how to plot US counties at all three available resolutions.


```python
import cartopy.crs as ccrs
import cartopy
import matplotlib.pyplot as plt

from metpy.plots import USCOUNTIES
from shapely.geometry import Point, Polygon
from cartopy.mpl.patch import geos_to_path

proj = ccrs.LambertConformal(central_longitude=-85, central_latitude=30.21)

counties_to_highlight = [(-92.033, 30.2167), (-90.82, 29.79), (-91.17, 30.39)] # lafayette, lafourche, br

fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(1, 3, 1, projection=proj)
ax1.set_extent([265, 272, 29, 34], ccrs.Geodetic())
ax1.add_feature(USCOUNTIES.with_scale('5m'))

center_county_color = 'black'  # Color for the county containing the center point
surrounding_county_color = 'blue'  # Color for the surrounding counties

for county in USCOUNTIES.with_scale('5m').geometries():
    for pt in counties_to_highlight:
        if county.contains(Point(pt)):
            ax1.add_geometries([county], ccrs.PlateCarree(), facecolor=center_county_color)


plt.show()
```
